````python3
# ì´ì§„ íƒìƒ‰ì˜ ì‹œê°ì  ì´í•´
print("=== ì´ì§„ íƒìƒ‰ ê³¼ì • ì‹œê°í™” ===")
print("\nì •ë‹µì´ 77ì¼ ë•Œ, 1~100 ë²”ìœ„ì—ì„œ ì°¾ëŠ” ê³¼ì •:\n")

secret = 77
left = 1
right = 100
attempt = 0

while left <= right:
    mid = (left + right) // 2
    attempt = attempt + 1
    
    print(f"ì‹œë„ {attempt}: ë²”ìœ„ [{left}-{right}] â†’ ì¤‘ê°„ê°’: {mid}", end="")
    
    if mid == secret:
        print(f" â†’ ì •ë‹µ!")
        print(f"\n{attempt}ë²ˆ ë§Œì— {secret}ì„(ë¥¼) ì°¾ì•˜ìŠµë‹ˆë‹¤!")
        break
    elif mid < secret:
        print(f" â†’ {secret}ì´(ê°€) {mid}ë³´ë‹¤ í¬ë‹¤ (UP)")
        left = mid + 1
    else:
        print(f" â†’ {secret}ì´(ê°€) {mid}ë³´ë‹¤ ì‘ë‹¤ (DOWN)")
        right = mid - 1
````

1. ì‚¬ìš©ìê°€ ë²”ìœ„ ì…ë ¥ (ìµœì†Ÿê°’ n, ìµœëŒ“ê°’ m)
2. ì‚¬ìš©ìê°€ ê·¸ ë²”ìœ„ ì•ˆì˜ ìˆ«ìë¥¼ ìƒê°í•¨
3. ì»´í“¨í„°ê°€ ì´ì§„ íƒìƒ‰ìœ¼ë¡œ ê·¸ ìˆ«ìë¥¼ ì°¾ìŒ

````python3
n = int(input("ìµœì†Ÿê°’ì„ ì…ë ¥í•˜ì„¸ìš”: "))
m = int(input("ìµœëŒ“ê°’ì„ ì…ë ¥í•˜ì„¸ìš”: "))
secret = int(input("ìƒê°í•œ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”: "))

left = n
right = m
attempts = 0

while left <= right:
    mid = (left + right) // 2
    attempts = attempts + 1
    
    print(f"ì‹œë„ {attempts}: [{left}-{right}], {mid}")
    
    if mid == secret:
        print(f"\nì •ë‹µì€ {mid}ì…ë‹ˆë‹¤!")
        break
    elif mid < secret:
        print(f"   â†’ UP! {mid}ë³´ë‹¤ í½ë‹ˆë‹¤.\n")
        left = mid + 1
    else:
        print(f"   â†’ DOWN! {mid}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤.\n")
        right = mid - 1
````
