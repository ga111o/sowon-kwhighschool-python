## 1. 조건문 (if-elif-else) 복습

조건문은 특정 조건에 따라 다른 코드를 실행하게 해줍니다.

**기본 형식:**
```python
if 조건1:
    # 조건1이 참일 때 실행
elif 조건2:
    # 조건2가 참일 때 실행
else:
    # 모든 조건이 거짓일 때 실행
```

````python
score = 85

if score >= 90:
    print("A등급입니다!")
elif score >= 80:
    print("B등급입니다!")
elif score >= 70:
    print("C등급입니다!")
else:
    print("더 노력이 필요합니다!")
````

````python
a = 10
b = 20

print(f"a == b: {a == b}")  # 같다
print(f"a != b: {a != b}")  # 다르다
print(f"a > b: {a > b}")    # 크다
print(f"a < b: {a < b}")    # 작다
print(f"a >= b: {a >= b}")  # 크거나 같다
print(f"a <= b: {a <= b}")  # 작거나 같다
````

**while 문:** 조건이 참인 동안 계속 반복
```python
while 조건:
    # 반복할 코드
```

**for 문:** 정해진 횟수만큼 반복
```python
for 변수 in range(횟수):
    # 반복할 코드
```

````python
# for 문 예제: 1부터 5까지 출력
for i in range(1, 6):
    print(f"{i}번째 반복")
````

````python
count = 5
while count > 0:
    print(f"{count}초 남았습니다!")
    count -= 1
print("시작!")
````

````python
for i in range(1, 11):
    if i == 5:
        print("5에서 멈춥니다!")
        break
    print(i)

for i in range(1, 6):
    if i == 3:
        print("3은 건너뜁니다!")
        continue
    print(i)
````

`random` 모듈을 사용하면 무작위 값을 생성할 수 있습니다.

주요 함수:
- `random.randint(a, b)`: a부터 b까지의 정수 중 무작위로 하나 선택
- `random.choice(리스트)`: 리스트에서 무작위로 하나 선택

````python
# random 모듈 import
import random

# randint 예제: 주사위 던지기
print("=== 주사위 던지기 ===")
dice = random.randint(1, 6)
print(f"주사위 결과: {dice}")

# 여러 번 던져보기
print("\n주사위를 5번 던집니다:")
for i in range(5):
    dice = random.randint(1, 6)
    print(f"{i+1}번째: {dice}")
````

````python
fruits = ["사과", "바나나", "딸기", "포도", "수박"]
selected = random.choice(fruits)
print(f"오늘의 과일: {selected}")
````

# 가위바위보 게임 만들기

이제 배운 내용을 활용해서 가위바위보 게임을 만들어봅시다!

요구사항
1. 사용자가 `가위`/`바위`/`보` 중 하나를 입력 (만약 그 외를 입력했을 때에는 "다시 입력해주세요"가 떠야 함)
2. 컴퓨터가 이겼을 경우 `컴퓨터 승리!`가 뜨고, 사용자가 이겼을 경우 `사용자 승리!`, 비겼을 경우, `비겼습니다`가 떠야 함
3. 사용자가 3번 연속 승리할 때까지 계속 반복
