````python
from bs4 import BeautifulSoup
import urllib.request # url lib은, 위에서 배운 request와 비슷한데, url을 쉽게 다룰 수 있도록 함.
import re

url = 'https://www.hollys.co.kr/store/korea/korStore2.do'

# urllib.request: urllib 모듈의 하위 모듈로, 웹 페이지에 HTTP 요청을 보내고 응답을 받을 수 있는 기능을 제공
# urlopen 지정된 URL에 HTTP 요청을 보내고 서버로부터 응답을 받아오는 함수, 변수에는 requests와 비슷한 형식의 객체가 담김.
html = urllib.request.urlopen(url)

soup = BeautifulSoup(html, 'html.parser')

# css path에 대해서 설명
tags = soup.select('html body div.wrap div#contentsWr.contents_wr div#contents.contents.po_design div.content fieldset fieldset div.tableType01 table.tb_store tbody tr td.center_t a')

store_numbers = []

for tag in tags:
    onclick_text = tag.get('onclick', '')

    # r'storeView\((\d+)\)': 이 패턴은 storeView(숫자) 형식을 찾기 위해 사용됨.
    match = re.search(r'storeView\((\d+)\)', onclick_text)
    if match:
        store_number = match.group(1)
        store_numbers.append(store_number)

print(store_numbers)
````

````python
# 중복되는 값이 있으니, set으로 중복값 제거.

unique_store_numbers = list(set(store_numbers))
print(unique_store_numbers)
````
